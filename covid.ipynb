```python
daily_cases = data.groupby('dateRep').sum()['cases']

plt.figure(figsize=(10,6))
daily_cases.plot(kind='line', color='b')
plt.title('Évolution des cas COVID-19 par jour')
plt.xlabel('Date')
plt.ylabel('Nombre de cas')
plt.grid(True)
plt.show()
```

```python

country_cases = data.groupby('countriesAndTerritories').sum(numeric_only=True)['cases']

country_cases = country_cases.sort_values(ascending=False)

country_cases.head(10)


country_cases.head(10).plot(kind='bar', figsize=(10,6))
plt.title('Top 10 des pays avec le plus grand nombre de cas COVID-19')
plt.xlabel('Pays')
plt.ylabel('Nombre de cas')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data['cases'], kde=True, color='blue')
plt.title('Distribution des cas COVID-19')
plt.xlabel('Nombre de cas')
plt.ylabel('Fréquence')
plt.show()
```

```python

data.groupby('continentExp')['cases'].sum().plot(kind='pie', figsize=(10, 6), autopct='%1.1f%%', startangle=90)
plt.title('Répartition des cas COVID-19 par continent')
plt.ylabel('')  # Supprimer l'étiquette de l'axe Y
plt.show()
```

```python

total_cases_by_continent = data.groupby('continentExp')['cases'].sum()

# Affichage sous forme de diagramme à barres
total_cases_by_continent.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Cas totaux de COVID-19 par continent')
plt.xlabel('Continent')
plt.ylabel('Cas')
plt.show()
```

```python

data['dateRep'] = pd.to_datetime(data['dateRep'])


cases_by_country = data.groupby(['countriesAndTerritories', 'dateRep'])['cases'].sum().reset_index()

cases_by_country = cases_by_country.pivot(index='countriesAndTerritories', columns='dateRep', values='cases')


# Afficher l'évolution des cas pour un pays spécifique (exemple: 'France')
cases_by_country.loc['France'].plot(figsize=(12, 6), title="Evolution des cas COVID-19 en France")  # Use .loc to access by index label
plt.xlabel('Date')
plt.ylabel('Cas')
plt.grid(True)
plt.show()
```
